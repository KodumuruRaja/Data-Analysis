```python
x=[1,2,3,4,5,6]
y=[10,20,30,40,50,60]
z=[100,200,300,400,500,600]
s=[1000,2000,3000,4000,5000,6000]
```

```python
plt.plot(x,y,label='Tens')
plt.plot(x,z,label='Hundreads')
plt.plot(x,s,label='Thousands')
plt.xlabel("x data points")
plt.ylabel("other data points")
plt.title("Plotting")
plt.legend()
```

```python
plt.plot(x,y,label='Tens')
plt.plot(x,z,label='Hundreads')
plt.plot(x,s,label='Thousands')
plt.xlabel("x data points")
plt.ylabel("other data points")
plt.title("Plotting")
plt.legend(loc='upper right')
```

```python
plt.plot(x,y,label='Tens')
plt.plot(x,z,label='Hundreads')
plt.plot(x,s,label='Thousands')
plt.xlabel("x data points")
plt.ylabel("other data points")
plt.title("Plotting")
plt.legend(loc='best',shadow=True)
```

```python
plt.plot(x,y,label='Tens')
plt.plot(x,z,label='Hundreads')
plt.plot(x,s,label='Thousands')
plt.xlabel("x data points")
plt.ylabel("other data points")
plt.title("Plotting")
plt.legend(loc='best',shadow=True,fontsize='large')
```

```python
plt.plot(x,y,label='Tens')
plt.plot(x,z,label='Hundreads')
plt.plot(x,s,label='Thousands')
plt.xlabel("x data points")
plt.ylabel("other data points")
plt.title("Plotting")
plt.legend(loc='best',shadow=True,fontsize='large')
plt.grid()
```

```python
t=['A','B','C','D','E','F']
plt.xticks(y,t)
plt.bar(y,z)
```

```python
p=['a','b','c','d','e','f']
plt.xticks(y,p)
plt.yticks(z,p)
plt.bar(y,z)
```

```python
plt.title("Plotting data points")
plt.xticks(y,p)
plt.xlabel("X data points")
plt.yticks(z,p)
plt.ylabel("Y data points")
plt.bar(y,z,label='Raja')
plt.legend()
```

```python
plt.title("Plotting data points")
plt.xticks(y,p)
plt.xlabel("X data points")
plt.yticks(z,p)
plt.ylabel("Y data points")
plt.bar(y,z,label='Raja')
plt.bar(x,y,label='Nivi')
plt.legend()
```

```python
plt.title("Plotting data points")
plt.xticks(y,p)
plt.xlabel("X data points")
plt.yticks(z,p)
plt.ylabel("Y data points")
plt.bar(y,z,width=3,label='Raja')
plt.bar(x,y,width=3,label='Nivi')
plt.legend()
```

```python
plt.title("Plotting data points")
plt.xticks(y,p)
plt.xlabel("X data points")
plt.yticks(z,p)
plt.ylabel("Y data points")
plt.barh(y,z,label='Raja')
plt.barh(x,y,label='Nivi')
plt.legend()
```
