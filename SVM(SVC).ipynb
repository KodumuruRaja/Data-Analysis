```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
```

```python
df=load_iris()
df
```

```python
data=pd.DataFrame(df.data,columns=df.feature_names)
data
```

```python
data['target']=df.target
data
```

```python
df.target_names
```

```python
data['flowers']=data.target.apply(lambda x: df.target_names[x])
data
```

```python
data0=data[data.target==0]
data1=data[data.target==1]
data2=data[data.target==2]
```

```python
data0
```

```python
data1
```

```python
plt.xlabel("Sepal Leangth (CM)")
plt.ylabel("Sepal Width (CM)")
plt.scatter(data0['sepal length (cm)'],data0['sepal width (cm)'],color='red',marker='o')
plt.scatter(data1['sepal length (cm)'],data1['sepal width (cm)'],color='yellow',marker='o')
```

```python
plt.xlabel("Petal Leangth (CM)")
plt.ylabel("Petal Width (CM)")
plt.scatter(data0['petal length (cm)'],data0['petal width (cm)'],color='red',marker='o')
plt.scatter(data1['petal length (cm)'],data1['petal width (cm)'],color='yellow',marker='o')
```

```python
x=data.drop(['target','flowers'],axis=1)
x
```

```python
y=data['target']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC 
```

```python
svm=SVC(kernel='linear')
```

```python
svm.fit(x_train,y_train)
```

```python
svm.score(x_test,y_test)
```
